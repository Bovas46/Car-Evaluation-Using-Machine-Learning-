This block reads the uploaded CSV file (car_evaluation.csv) into a pandas DataFrame and gives thoughtful names to the columns. It showcases the first few rows of the dataset,glimpse of the features (like safety and buying price) and the target label (class) to help you get a sense of the data.

```python
#Block 2
import pandas as pd

# Load the uploaded CSV file
df = pd.read_csv('/content/car_evaluation.csv', header=None)
df.columns = ["buying", "maint", "doors", "persons", "lug_boot", "safety", "class"]

# Display first few rows
df.head()
```

This block explores the dataset to detect missing values, understand feature distributions, and visualize class balance. It also uses sample pairplots to observe relationships between input features and the target class, which helps assess potential patterns in the data before model building.

Here, all categorical features are converted into numeric values using LabelEncoder, a necessary step because machine learning models require numerical inputs. The block also separates the features (X) from the target variable (y), preparing the dataset for training and evaluation

```python
#Block 4
from sklearn.preprocessing import LabelEncoder

# Encode all categorical columns
label_encoders = {}
for column in df.columns:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

# Features and target
X = df.drop("class", axis=1)
y = df["class"]
```

This block splits the dataset into training and test sets using a friendly 70/30 ratio. This is super important to see how well each model can handle new, unseen data. Setting a random seed makes sure the same split can be recreated for consistent results, which is just wonderful.

```python
#Block 5
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

In this block, we train a Logistic Regression model using the training data and then put it to the test on the unseen test data. The classification_report function checks out the model's accuracy, precision, recall, and F1-score, giving us a good understanding of how well it classifies car evaluation categories.

```python
#Block 6
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)

print("Logistic Regression Results:\n")
print(classification_report(y_test, y_pred_lr))
```

This block brings the k-Nearest Neighbors algorithm to life with k=5, fits it to the training data, and makes predictions on the test set. It assesses performance using standard classification metrics, making it easy to compare with other algorithms based on their predictive accuracy and other scores.

```python
#Block 7
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

print("k-NN Results:\n")
print(classification_report(y_test, y_pred_knn))
```

This block builds a Decision Tree classifier to model the data. It trains on the training subset and predicts on the test set. The results, printed as a classification report, help determine how well the decision tree performs in identifying car categories correctly.

```python
#Block 8
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

print("Decision Tree Results:\n")
print(classification_report(y_test, y_pred_dt))
```

This block calculates and visualizes performance metrics (accuracy, precision, recall, F1-score) for all three models. Using delightful bar plots, it makes visual comparison across models a breeze, helping us pinpoint the best-performing algorithm based on specific evaluation criteria and enhancing interpretation.

```python
#Block 9
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def get_metrics(y_true, y_pred):
    return [
        accuracy_score(y_true, y_pred),
        precision_score(y_true, y_pred, average='weighted'),
        recall_score(y_true, y_pred, average='weighted'),
        f1_score(y_true, y_pred, average='weighted')
    ]

models = ["Logistic Regression", "k-NN", "Decision Tree"]
metrics = np.array([
    get_metrics(y_test, y_pred_lr),
    get_metrics(y_test, y_pred_knn),
    get_metrics(y_test, y_pred_dt)
])

metric_names = ["Accuracy", "Precision", "Recall", "F1 Score"]

for i, name in enumerate(metric_names):
    plt.figure()
    plt.bar(models, metrics[:, i], color=["skyblue", "orange", "lightgreen"])
    plt.title(name + " Comparison")
    plt.ylabel(name)
    plt.ylim(0, 1)
    plt.show()
```

This block analyzes and compares the performance of Logistic Regression, k-NN, and Decision Tree classifiers. It computes evaluation metrics and presents them in both numerical and visual forms. Using descriptive stats and plots, it highlights the best-performing model under each criterion and helps identify the most reliable algorithm for this dataset.

```python
#Block 11
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Collect predictions for all models
models = {
    "Logistic Regression": y_pred_lr,
    "k-Nearest Neighbors": y_pred_knn,
    "Decision Tree": y_pred_dt
}

# Evaluate each model
comparison_metrics = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],
    "Recall": [],
    "F1-Score": []
}

for name, preds in models.items():
    comparison_metrics["Model"].append(name)
    comparison_metrics["Accuracy"].append(accuracy_score(y_test, preds))
    comparison_metrics["Precision"].append(precision_score(y_test, preds, average='weighted', zero_division=0))
    comparison_metrics["Recall"].append(recall_score(y_test, preds, average='weighted'))
    comparison_metrics["F1-Score"].append(f1_score(y_test, preds, average='weighted'))

# Convert to DataFrame
metrics_df = pd.DataFrame(comparison_metrics)

# Descriptive summary
print("=== Descriptive Statistics for Model Metrics ===")
print(metrics_df.describe())

# Best-performing model per metric
best_models = metrics_df.set_index("Model").idxmax()
print("\n=== Best Model for Each Metric ===")
print(best_models)

# Visual comparison using Seaborn
plt.figure(figsize=(10, 6))
melted_df = metrics_df.melt(id_vars="Model", var_name="Metric", value_name="Score")
sns.barplot(data=melted_df, x="Metric", y="Score", hue="Model", palette="muted")
plt.title("Model Comparison on Evaluation Metrics")
plt.ylim(0, 1.05)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
